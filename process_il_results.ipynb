```python
import wandb
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
api = wandb.Api()
run = api.run("willschwarzer/sirl_evaluation/1b5q4hod")
run_percentiles = api.run("willschwarzer/sirl_evaluation/3e51gwlj")
```

```python
data = run.history(samples=1000000)

data
# filtered_df = orig[orig['rl_pred'].notnull()][:62]
```

```python
data_percentiles = run_percentiles.history(samples=1000000)
data_percentiles
```

```python
data_percentiles['ave_gail_perc'] = np.mean(np.array([perc for perc in data_percentiles['gail_percentiles']]), axis=1)
data_percentiles['ave_gt_perc'] = np.mean(np.array([perc for perc in data_percentiles['ground_truth_percentiles']]), axis=1)
data_percentiles['ave_sirl_perc'] = np.mean(np.array([perc for perc in data_percentiles['sirl_percentiles']]), axis=1)
data_percentiles['ave_bc_perc'] = np.mean(np.array([perc for perc in data_percentiles['bc_percentiles']]), axis=1)
data_percentiles['ave_airl_perc'] = np.mean(np.array([perc for perc in data_percentiles['airl_percentiles']]), axis=1)
```

```python
data_percentiles[['ave_gt_perc', 'ave_sirl_perc', 'ave_bc_perc', 'ave_gail_perc', 'ave_airl_perc']]
```

```python
print('Average GAIL Percentiles:', np.mean(np.array([perc for perc in data_percentiles['gail_percentiles']]), axis=0))
```

```python
data["sirl_improvement"] = (data["sirl_ret"] - data["random_rew_ret"])/(data["ground_truth_ret"] - data["random_rew_ret"])
data["gail_improvement"] = (data["gail_ret"] - data["random_rew_ret"])/(data["ground_truth_ret"] - data["random_rew_ret"])
data["airl_improvement"] = (data["airl_ret"] - data["random_rew_ret"])/(data["ground_truth_ret"] - data["random_rew_ret"])
```

```python
data
```

```python
print("SIRL:", np.mean(data.sirl_improvement))
print("GAIL:", np.mean(data.gail_improvement))
print("AIRL:", np.mean(data.airl_improvement))
```

```python
bins = np.array([-0.5, 0, 0.5, 1])
plt.hist((data.sirl_improvement, data.gail_improvement, data.airl_improvement), label=["sirl", "gail", "airl"], bins=bins)
plt.legend()
```
