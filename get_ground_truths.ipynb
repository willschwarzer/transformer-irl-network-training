```python
import torch
import numpy as np
from utils import get_freest_gpu
import torch.nn.functional as F
from new_block_env_new import NewBlockEnv
import os
```

```python
def string_to_rewards(st):
    return [float(num) for num in st.split("_")]
```

```python
models_directory = "models_regression"
models = os.listdir(models_directory)
```

```python
models
```

```python
num_models = len(models)
```

```python
all_weights = np.zeros((num_models, 3), dtype=float)
```

```python
print(f"Found {num_models} models")
for idx, model_name in enumerate(models):
    weights = string_to_rewards(model_name)
    print(weights)
    all_weights[idx] = weights
```

```python
np.save(f"ground_truth_{num_models}.npy", all_weights)
```

```python
np.save(f"ground_truth_101.npy", all_weights[:101])
```
